```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/placement (1).csv')
df.head()
```

```python
plt.figure(figsize=(16,5))
plt.subplot(121)
sns.distplot(df['cgpa'])
plt.subplot(122)
sns.distplot(df['placement_exam_marks'])
plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.subplot(121)
sns.distplot(df['cgpa'])
plt.subplot(122)
sns.distplot(df['placement_exam_marks'])
plt.show()
```

This method is used when the data is right skewed or left skewed

```python
df['placement_exam_marks'].describe()
```

```python
p75=df['placement_exam_marks'].quantile(0.75)
p25=df['placement_exam_marks'].quantile(0.25)
```

```python
iqr=p75-p25
```

```python
ul=p75+1.5*iqr
ll=p25-1.5*iqr
```

Trimming

```python
new_df=df[(df['placement_exam_marks']>ll)&(df['placement_exam_marks']<ul)]
new_df
```

```python
plt.figure(figsize=(16,8))
plt.subplot(221)
sns.distplot(df['placement_exam_marks'])
plt.subplot(222)
sns.boxplot(df['placement_exam_marks'])
plt.subplot(223)
sns.distplot(new_df['placement_exam_marks'])
plt.subplot(224)
sns.boxplot(new_df['placement_exam_marks'])
plt.show()
```

Capping

```python
df['placement_exam_marks']=np.where(df['placement_exam_marks']>=ul,ul,np.where(df['placement_exam_marks']<ll,ll,df['placement_exam_marks']))
```

```python
plt.figure(figsize=(16,7))
plt.subplot(221)
sns.distplot(df['placement_exam_marks'])
plt.subplot(222)
sns.boxplot(df['placement_exam_marks'])
plt.show()
```
