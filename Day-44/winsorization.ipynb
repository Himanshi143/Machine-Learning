```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/weight-height.csv')
df.head()
```

```python
plt.figure(figsize=(16,7))
plt.subplot(221)
sns.distplot(df['Height'])
plt.subplot(222)
sns.boxplot(df['Height'])
plt.show()
```

```python
df['Height'].describe()
```

```python
ul=df['Height'].quantile(0.99)
ll=df['Height'].quantile(0.01)
```

Trimming

```python
new_df=df[(df['Height']>ll)&(df['Height']<ul)]
new_df
```

```python
plt.figure(figsize=(16,8))
plt.subplot(221)
sns.distplot(df['Height'])
plt.subplot(222)
sns.boxplot(df['Height'])
plt.subplot(223)
sns.distplot(new_df['Height'])
plt.subplot(224)
sns.boxplot(new_df['Height'])
plt.show()
```

Capping-winsorization

```python
df['Height']=np.where(df['Height']>=ul,ul,np.where(df['Height']<ll,ll,df['Height']))
```

```python
plt.figure(figsize=(16,7))
plt.subplot(221)
sns.distplot(df['Height'])
plt.subplot(222)
sns.boxplot(df['Height'])
plt.show()
```
