```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('/content/wine_data (1).csv',usecols=[0,1,2],header=None,names=['Class label', 'Alcohol', 'Malic acid'])
df.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop('Class label',axis=1),df['Class label'],test_size=0.2)
x_train
```

```python
mms=MinMaxScaler()
x_train_mms=mms.fit_transform(x_train)
x_test_mms=mms.transform(x_test)
x_train_mms=pd.DataFrame(x_train_mms,columns=x_train.columns)
x_test_mms=pd.DataFrame(x_test_mms,columns=x_train.columns)
```

```python
sns.relplot(data=x_train,kind='scatter',x='Alcohol',y='Malic acid')
sns.relplot(data=x_train_mms,kind='scatter',x='Alcohol',y='Malic acid')
```

```python
sns.displot(data=x_train,kind='kde',x='Alcohol',y='Malic acid')
sns.displot(data=x_train_mms,kind='kde',x='Alcohol',y='Malic acid')
```

Distributions before and after scaling
