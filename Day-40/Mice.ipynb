```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv('/content/50_Startups.csv',usecols=[0,1,2])
df=np.round(df/10000,2)
df=df.sample(5)
df.head()
```

```python
df.iloc[1,0] = np.NaN
df.iloc[3,1] = np.NaN
df.iloc[-1,-1] = np.NaN
```

```python
df
```

```python
si=SimpleImputer()
df_train=si.fit_transform(df)
```

```python
df0=pd.DataFrame(np.round(df_train,2),columns=df.columns)
df0
```

```python
df1=df0.copy()
df1.iloc[1,0]=np.NaN
df1
```

```python
lr=LinearRegression()
lr.fit(df1.iloc[[0,2,3,4],1:],df1.iloc[[0,2,3,4],:1])
lr.predict(df1.iloc[1,1:].values.reshape(1,2))
```

```python
df1.iloc[1,0]=22.97
df1
```

```python
df1.iloc[3,1]=np.NaN
print(df1)
lr=LinearRegression()
lr.fit(df1.iloc[[0,1,2,4],[0,2]],df1.iloc[[0,1,2,4],[1]])
df1.iloc[3,1]=np.round(lr.predict(df1.iloc[3,[0,2]].values.reshape(1,2)),2)
df1
```

```python
df1.iloc[-1,-1]=np.NaN
print(df1)
lr=LinearRegression()
lr.fit(df1.iloc[[0,1,2,3],[0,1]],df1.iloc[[0,1,2,3],[2]])
df1.iloc[-1,-1]=np.round(lr.predict(df1.iloc[4,[0,1]].values.reshape(1,2)),2)
df1
```

```python
df1-df0
```

```python
df2=df1.copy()
df2.iloc[1,0]=np.NaN
print(df2)
lr=LinearRegression()
lr.fit(df2.iloc[[0,2,3,4],1:],df2.iloc[[0,2,3,4],:1])
df2.iloc[1,0]=lr.predict(df2.iloc[1,1:].values.reshape(1,2))
df2
```

```python
df2.iloc[3,1]=np.NaN
print(df2)
lr=LinearRegression()
lr.fit(df2.iloc[[0,1,2,4],[0,2]],df2.iloc[[0,1,2,4],[1]])
df2.iloc[3,1]=np.round(lr.predict(df2.iloc[3,[0,2]].values.reshape(1,2)),2)
df2
```

```python
df2.iloc[-1,-1]=np.NaN
print(df2)
lr=LinearRegression()
lr.fit(df2.iloc[[0,1,2,3],[0,1]],df2.iloc[[0,1,2,3],[2]])
df2.iloc[-1,-1]=np.round(lr.predict(df2.iloc[4,[0,1]].values.reshape(1,2)),2)
df2
```
