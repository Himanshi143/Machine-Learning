```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.impute import KNNImputer,SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('/content/train (1).csv',usecols=['Age','Pclass','Fare','Survived'])
df.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop('Survived',axis=1),df['Survived'],test_size=0.2)
x_train
```

```python
knn=KNNImputer()
x_train_trf=knn.fit_transform(x_train)
x_test_trf=knn.fit_transform(x_test)
```

```python
lr=LogisticRegression()
lr.fit(x_train_trf,y_train)
```

```python
y_pred=lr.predict(x_test_trf)
```

```python
accuracy_score(y_test,y_pred)
```

```python
si=SimpleImputer()
x_train_si=si.fit_transform(x_train)
x_test_si=si.fit_transform(x_test)
```

```python
lr=LogisticRegression()
lr.fit(x_train_si,y_train)
```

```python
y_pred_si=lr.predict(x_test_si)
```

```python
accuracy_score(y_test,y_pred_si)
```
