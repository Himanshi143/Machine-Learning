```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('/content/Social_Network_Ads (3).csv',usecols=['Age','EstimatedSalary','Purchased'])
df.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,:2],df.iloc[:,2:],test_size=0.2)
```

```python
ss=StandardScaler()
x_train_ss=ss.fit_transform(x_train)
x_test_ss=ss.transform(x_test)
```

```python
x_train_ss=pd.DataFrame(x_train_ss,columns=x_train.columns)
x_test_ss=pd.DataFrame(x_test_ss,columns=x_train.columns)
```

Why scaling is important?

```python
lr=LogisticRegression()
lr_ss=LogisticRegression()
lr.fit(x_train,y_train)
lr_ss.fit(x_train_ss,y_train)
y_pred=lr.predict(x_test)
y_pred_ss=lr_ss.predict(x_test_ss)
```

```python
print(accuracy_score(y_test,y_pred))
print(accuracy_score(y_test,y_pred_ss))
```

```python
dtc=DecisionTreeClassifier()
dtc_ss=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
dtc_ss.fit(x_train_ss,y_train)
y_pred=dtc.predict(x_test)
y_pred_ss=dtc_ss.predict(x_test_ss)
```

```python
print(accuracy_score(y_test,y_pred))
print(accuracy_score(y_test,y_pred_ss))
```

Effect of Outlier

```python
df1=pd.DataFrame({'Age':[5,90,95],'EstimatedSalary':[1000,250000,350000],'Purchased':[0,1,1]})
df=pd.concat([df,df1])
```

```python
plt.scatter(df['Age'], df['EstimatedSalary'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased', axis=1),
                                                    df['Purchased'],
                                                    test_size=0.3,
                                                    random_state=0)
```

```python
scaler = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train['Age'], X_train['EstimatedSalary'])
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'],color='red')
ax2.set_title("After Scaling")
plt.show()
```
