```python
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('/content/train.csv',usecols=[1,2,5,6,7])
df.head()
```

```python
df.dropna(inplace=True)
```

```python
x=df.iloc[:,1:]
y=df.iloc[:,0]
```

```python
x
```

```python
y
```

```python
np.mean(cross_val_score(LogisticRegression(),x,y,scoring='accuracy',cv=20))
```

```python
df['family_size']=df['Parch']+df['SibSp']
```

```python
df.head()
```

```python
def fam(num):
  if(num==0):
    return 0
  elif(num>0 and num<=4):
    return 1
  else:
    return 2
```

```python
df['fam_type']=df['family_size'].apply(fam)
```

```python
df.head()
```

```python
df.drop(columns=['SibSp','Parch','family_size'],inplace=True)
```

```python
df.head()
```

```python
x_=df.iloc[:,1:]
y_=df.iloc[:,0]
```

```python
np.mean(cross_val_score(LogisticRegression(),x_,y_,scoring='accuracy',cv=20))
```
